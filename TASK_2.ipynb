```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.lib.pagesizes import A4
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path = "/content/Car_sales.csv"
```

```python
data = pd.read_csv("/content/Car_sales.csv")
```

```python
#head will give starting 5 value
data.head()
```

```python
# tail will give last 5 value
data.tail()
```

```python
#  it shows how many row and column do we have
data.shape
```

```python

#it describe the data into Q1, Q2 and min and maximum value
summary=data.describe()
```

```python
data.info()
```

```python
#checking null value
null_counts=data.isnull().sum()
```

```python
# drop null value
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
#cheching duplicated value
data.duplicated().sum()
```

```python
#  3. Create PDF Report
# ------------------------------
pdf_file = "Data_Analysis_Report.pdf"
doc = SimpleDocTemplate(pdf_file, pagesize=A4)
styles = getSampleStyleSheet()
elements = []

# Title
elements.append(Paragraph("📊 Data Analysis Report", styles["Title"]))
elements.append(Spacer(1, 12))

# File Info
elements.append(Paragraph(f"File Analyzed: {file_path}", styles["Normal"]))
elements.append(Paragraph(f"Total Rows: {len(data)}, Total Columns: {data.shape[1]}", styles["Normal"]))
elements.append(Spacer(1, 12))

# Summary Statistics Table
elements.append(Paragraph("Summary Statistics", styles["Heading2"]))
summary_table_data = [summary.reset_index().columns.tolist()] + summary.reset_index().values.tolist()
summary_table = Table(summary_table_data)
summary_table.setStyle(TableStyle([
    ("BACKGROUND", (0, 0), (-1, 0), colors.gray),
    ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
    ("ALIGN", (0, 0), (-1, -1), "CENTER"),
    ("GRID", (0, 0), (-1, -1), 0.5, colors.black)
]))
elements.append(summary_table)
elements.append(Spacer(1, 12))

# Missing Values Table
elements.append(Paragraph("Missing Values", styles["Heading2"]))
# Convert the Series to a DataFrame before creating the table data
null_table_data = [null_counts.reset_index().columns.tolist()] + null_counts.reset_index().values.tolist()
null_table = Table(null_table_data)
null_table.setStyle(TableStyle([
    ("BACKGROUND", (0, 0), (-1, 0), colors.darkblue),
    ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
    ("ALIGN", (0, 0), (-1, -1), "CENTER"),
    ("GRID", (0, 0), (-1, -1), 0.5, colors.black)
]))
elements.append(null_table)

# Save PDF
doc.build(elements)
print(f"✅ PDF Report saved as {pdf_file}")
```
